```python
def load_dataset(df):
    X = df.drop("class", axis=1)
    y = df["class"]
    return X, y
```

```python
X, y = load_dataset(df)
```

```python
# compare probabilistic models on the Haberman dataset
from numpy import mean, std
from pandas import read_csv
from matplotlib import pyplot
from sklearn.dummy import DummyClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from imblearn.pipeline import Pipeline as ImbPipeline


def evaluate(X, y, model):

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
    return scores


def build_pipeline(X, model):

    num_cols = X.select_dtypes(include=["int64", "float64"]).columns
    cat_cols = X.select_dtypes(include=["object", "category"]).columns

    column_tr = ColumnTransformer(
        transformers=[
            ("num", Pipeline([
                ("imputer", SimpleImputer(strategy="mean")),
                ("scaler", StandardScaler())
            ]), num_cols),
            ("cat", Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]), cat_cols)
        ]
    )

    pipeline = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', model)
    ])

    return pipeline
```

```python
pipe = build_pipeline(X,  DummyClassifier(strategy='stratified'))
```

```python
scores = evaluate(X, y, pipe)
print('Mean ROC AUC: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline as ImbPipeline


def retrieve_models(X):

    num_cols = X.select_dtypes(include=["int64", "float64"]).columns
    cat_cols = X.select_dtypes(include=["object", "category"]).columns

    column_tr = ColumnTransformer(
        transformers=[
            ("num", Pipeline([
                ("imputer", SimpleImputer(strategy="mean")),
                ("scaler", StandardScaler())
            ]), num_cols),
            ("cat", Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]), cat_cols)
        ]
    )

    models = dict()
    models['SVC'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', SVC(random_state=42))
    ])

    models['LR'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', LogisticRegression(max_iter=1000, random_state=42, solver="lbfgs"))
    ])

    # Discriminantes
    models['LDA'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', LinearDiscriminantAnalysis())
    ])

    models['QDA'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', QuadraticDiscriminantAnalysis())
    ])

    # Naive Bayes
    models['GNB'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),  # GNB tolera escala; StandardScaler suele ayudar
        ('model', GaussianNB())
    ])


    # Árboles / ensembles (invariantes a escala, pero no estorba llevar todo en el mismo pipeline)
    models['RF'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', RandomForestClassifier(random_state=42, n_estimators=1000))
    ])

    models['GB'] = ImbPipeline(steps=[
        ('preprocessor', column_tr),
        ('model', GradientBoostingClassifier(random_state=42))
    ])

    return models
```

```python
models = retrieve_models(X)
```

```python
results = list()

for key, value in models.items():
    scores = evaluate(X, y, value)
    results.append(scores)
    print(f"{key}: {mean(scores)} {std(scores)}")

pyplot.boxplot(results, labels=models.keys(), showmeans=True)
pyplot.show()
```

```python
from imblearn.under_sampling import (
    TomekLinks,
    EditedNearestNeighbours,
    RepeatedEditedNearestNeighbours,
    OneSidedSelection,
    NeighbourhoodCleaningRule
)
from imblearn.over_sampling import (
    RandomOverSampler,
    SMOTE,
    BorderlineSMOTE,
    ADASYN
)

def retrieve_undersampling_oversampling_techniques():
    techniques = {
        # =======================
        # UNDERSAMPLING
        # =======================
        "TomekLinks": TomekLinks(),
        "ENN": EditedNearestNeighbours(),
        "RENN": RepeatedEditedNearestNeighbours(),
        "OSS": OneSidedSelection(),
        "NCR": NeighbourhoodCleaningRule(),

        # =======================
        # OVERSAMPLING
        # =======================
        "RandomOverSampler": RandomOverSampler(),
        "SMOTE": SMOTE(),
        "BorderlineSMOTE": BorderlineSMOTE(),
        "ADASYN": ADASYN(),
    }

    return techniques

techniques = retrieve_undersampling_oversampling_techniques()
```

```python
from sklearn.preprocessing import FunctionTransformer
from numpy import sqrt

results = []

for key, value in techniques.items():
    num_cols = X.select_dtypes(include=["int64", "float64"]).columns
    cat_cols = X.select_dtypes(include=["object", "category"]).columns

    column_tr = ColumnTransformer(
        transformers=[
            ("num", Pipeline([
                ("imputer", SimpleImputer(strategy="mean")),
                ("sqrt", FunctionTransformer(sqrt, validate=False)),
                ("scaler", StandardScaler())
            ]), num_cols),
            ("cat", Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]), cat_cols)
        ]
    )
    pipeline = ImbPipeline([
        ('prep', column_tr),
        ('technique', value),
        ('model', RandomForestClassifier(random_state=42, n_estimators=100))
    ])
    scores = evaluate(X, y, pipeline)
    results.append(scores)
    print('>%s %.3f (%.3f)' % (key, mean(scores), std(scores)))

pyplot.boxplot(results, labels=techniques.keys(), showmeans=True)
pyplot.show()
```

```python
from sklearn.calibration import CalibratedClassifierCV


def build_final_pipeline(X):

    num_cols = X.select_dtypes(include=["int64", "float64"]).columns
    cat_cols = X.select_dtypes(include=["object", "category"]).columns

    column_tr = ColumnTransformer(
        transformers=[
            ("num", Pipeline([
                ("imputer", SimpleImputer(strategy="mean")),
                ("sqrt", FunctionTransformer(sqrt, validate=False)),
                ("scaler", StandardScaler())
            ]), num_cols),
            ("cat", Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]), cat_cols)
        ]
    )

    calibrated_model = CalibratedClassifierCV(estimator=RandomForestClassifier(random_state=42, n_estimators=1000, class_weight="balanced"), method="isotonic", cv=3)

    pipeline = ImbPipeline([
        ('prep', column_tr),
        ('technique', techniques['BorderlineSMOTE']),
        ('model', calibrated_model)
    ])

    return pipeline
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.4,
    random_state=42,
    stratify=y
)
```

```python
pipe = build_final_pipeline(X_train)
pipe.fit(X_train, y_train)
```

```python
import numpy as np

def get_y_scores(model, X):

    if hasattr(model, "predict_proba"):
        return model.predict_proba(X)[:, 1]
    if hasattr(model, "decision_function"):
        return model.decision_function(X)

    return model.predict(X)
```

```python
from sklearn.metrics import (
    roc_auc_score, roc_curve,
    precision_recall_curve, average_precision_score, auc
)
import matplotlib.pyplot as plt

def evaluate_on_test(model, X_train, y_train, X_test, y_test, plot=True):
    # Entrena
    model.fit(X_train, y_train)

    # Puntajes
    y_scores = get_y_scores(model, X_test)

    # ROC AUC
    roc_auc = roc_auc_score(y_test, y_scores)
    fpr, tpr, _ = roc_curve(y_test, y_scores)

    # Precision-Recall: AP (Average Precision) y PR-AUC
    precision, recall, _ = precision_recall_curve(y_test, y_scores)
    ap = average_precision_score(y_test, y_scores)  # (promedio ponderado de precision en todos los recalls)
    pr_auc = auc(recall, precision)                 # Área geométrica bajo la curva P-R

    if plot:
        # Curva ROC
        plt.figure()
        plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.3f}")
        plt.plot([0,1], [0,1], linestyle="--")
        plt.xlabel("FPR")
        plt.ylabel("TPR")
        plt.title("ROC Curve (test)")
        plt.legend()
        plt.show()

        # Curva Precision-Recall
        plt.figure()
        plt.plot(recall, precision, label=f"AP = {ap:.3f} | PR-AUC = {pr_auc:.3f}")
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.title("Precision-Recall Curve (test)")
        plt.legend()
        plt.show()

    return {"roc_auc": roc_auc, "average_precision": ap, "pr_auc": pr_auc}
```

```python
test_metrics = evaluate_on_test(pipe, X_train, y_train, X_test, y_test, plot=True)
print(test_metrics)
```
